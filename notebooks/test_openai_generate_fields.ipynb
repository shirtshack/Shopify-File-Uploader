```python
#import os # Uncomment this line if you don't use previous code chunk
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
product_name = "BIKER SKULL"
description = "a bicycle skull"

few_shot = f"""Product Description: A funny t-shirt called "AMONG US TRUST NO ONE" it shows many colorful aliens
Product Title: Among us trust no one, funny T-shirt, 100% Cotton, Mens T Shirt Funny, Unisex Printed Design Ideal birthday gift for true players. Who is the impostor? Find out!
Product Description: A funny t-shirt called "BEER THERAPY" it showsa man holding a beer
Product Title: Beer Therapy T-Shirt for Men - Funny and Refreshing! 100% cotton, Mens T Shirt Funny, Unisex Printed Design. The perfect therapy session after a long day. Cheers to that!
Product Description: A funny t-shirt called "ALL MIGHT SMASH" it shows hero academia
Product Title:All Might Smash T-Shirt - Funny and Powerful! Unleash your inner hero with this 100% cotton, Mens T Shirt Funny, Unisex Printed Design. Show the world your heroic spirit! Plus Ultra!
Product Description: A funny t-shirt called "{product_name}" it shows {description}
Product Title:"""

response = client.chat.completions.create(model="gpt-4o-mini", #model="gpt-3.5-turbo",
messages=[{"role": "system", "content": "You are a system that is trained to generate 250 character product titles for Amazon products. You are asked to generate a product title for the given product description. Add funny and apporpriate bits to make the title more interesting."}, 
{"role": "user", "content": few_shot}],

temperature=0.9,
max_tokens=256)
```

```python
response.choices[0].message.content
```

```python
#Generate keyphrases
product_name = "BIKER SKULL"
description = "a bicycle skull"

few_shot = f"""Please generate a list of 5 keyphrases for the following product: A funny t-shirt model named "{product_name}" that shows {description}."""

response = client.chat.completions.create(model="gpt-3.5-turbo",
messages=[{"role": "system", "content": "You are a keyphrase generation system that is trained to produce EXACTLY 5 keyphrases for Amazon products. Keywords should be relevant to the product and should be popular. The list of keyphrases should be comma separated"}, 
{"role": "user", "content": few_shot}],

temperature=0.9,
max_tokens=256)
response.choices[0].message.content
```
